```python
#Install libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# import dataframe_image as dfi
```

```python
#Reading in the csv files
df1= pd.read_csv('Prem Data/2021_2022.csv')
df2= pd.read_csv('Prem Data/2020_2021.csv')
df3= pd.read_csv('Prem Data/2019_2020.csv')
df4= pd.read_csv('Prem Data/2018_2019.csv')
df5= pd.read_csv('Prem Data/2017_2018.csv')
df6= pd.read_csv('Prem Data/2016_2017.csv')
df7= pd.read_csv('Prem Data/2015_2016.csv')
df8= pd.read_csv('Prem Data/2014_2015.csv')
df9= pd.read_csv('Prem Data/2013_2014.csv')
df10= pd.read_csv('Prem Data/2012_2013.csv')
df11= pd.read_csv('Prem Data/2011_2012.csv')
df12= pd.read_csv('Prem Data/2010_2011.csv')
df13= pd.read_csv('Prem Data/2009_2010.csv')
df14= pd.read_csv('Prem Data/2008_2009.csv')
```

```python
#Creating a new column called season in the dataframes
df1['Season'] = np.where(df1['Div']== 'E0', '21/22', '0' )
df2['Season'] = np.where(df2['Div']== 'E0', '20/21', '0' )
df3['Season'] = np.where(df3['Div']== 'E0', '19/20', '0' )
df4['Season'] = np.where(df4['Div']== 'E0', '18/19', '0' )
df5['Season'] = np.where(df5['Div']== 'E0', '17/18', '0' )
df6['Season'] = np.where(df6['Div']== 'E0', '16/17', '0' )
df7['Season'] = np.where(df7['Div']== 'E0', '15/16', '0' )
df8['Season'] = np.where(df8['Div']== 'E0', '14/15', '0' )
df9['Season'] = np.where(df9['Div']== 'E0', '13/14', '0' )
df10['Season'] = np.where(df10['Div']== 'E0', '12/13', '0' )
df11['Season'] = np.where(df11['Div']== 'E0', '11/12', '0' )
df12['Season'] = np.where(df12['Div']== 'E0', '10/11', '0' )
df13['Season'] = np.where(df13['Div']== 'E0', '09/10', '0' )
df14['Season'] = np.where(df14['Div']== 'E0', '08/09', '0' )
df13
```

```python
#Concatenating the separate dataframes into one
combined_df = pd.concat([df1, df2,df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14])
combined_df
```

```python
#showing the columns in the dataframe
print(combined_df.columns.tolist())
```

```python
#Filtering out the columns that well use in analysis. Then renaming the columns
columns_combined_df= combined_df[['Season', 'Time', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG','HTR',
            'Referee', 'HS', 'AS','HST', 'AST', 'HC',
              'AC', 'HF', 'AF','HY', 'AY', 'HR', 'AR']]
final_combined_df=columns_combined_df.rename(columns={'FTHG':'FT_home_goals', 'FTAG':'FT_away_goals', 'FTR':'FT_results', 
                                                   'HTHG':'HT_home_goals', 'HTAG':'HT_away_goals','HTR':'HT_results', 
                                                   'HS':'home_shots','AS':'away_shots','HST':'home_shots_on_target', 
                                                   'AST':'away_shots_on_targets', 'HC':'home_corners','AC':'away_corners', 
                                                   'HF':'home_fouls', 'AF':'away_fouls','HY':'home_yellow_cards', 
                                                   'AY':'away_yellow_cards', 'HR':'home_red_cards', 'AR':'away_red_cards'})

other_team_df=final_combined_df
other_team_df
final_combined_df
```

```python
print(final_combined_df.columns.tolist())
```

```python
#Filter the data for home records of the selected six teams
home_teams= final_combined_df.loc[(final_combined_df['HomeTeam']=='Arsenal') | (final_combined_df['HomeTeam']=='Chelsea') 
                                  | (final_combined_df['HomeTeam']=='Liverpool') |
                                  (final_combined_df['HomeTeam']=='Tottenham') | (final_combined_df['HomeTeam']=='Man City')
                                  | (final_combined_df['HomeTeam']=='Man United')]
#Filter the data for away records for the selected six teams
away_teams= final_combined_df.loc[(final_combined_df['AwayTeam']=='Arsenal') | (final_combined_df['AwayTeam']=='Chelsea') |
                                  (final_combined_df['AwayTeam']=='Liverpool') | 
                                  (final_combined_df['AwayTeam']=='Tottenham') |(final_combined_df['AwayTeam']=='Man City') 
                                  |(final_combined_df['AwayTeam']=='Man United')]
#home_teams
away_teams
```

## GOALS (big six)

```python
# Number of goals scored each season when the big six played away or at home
homet_season_goals= home_teams.groupby('Season').sum()
awayt_season_goals= away_teams.groupby('Season').sum()
awayt_season_goals
```

```python
#Do the above but for all other teams that werent big six so we can 
#get a ration of goals scored by these top teams in relation to the rest of the league
del_teams= ['Arsenal', 'Chelsea', 'Liverpool', 'Tottenham', 'Man City', 'Man United']
other_hteams= other_team_df.loc[~other_team_df['HomeTeam'].isin(del_teams)]
other_ateams= other_team_df.loc[~other_team_df['AwayTeam'].isin(del_teams)]

#group by season and aggregate by sum to get the total number of goals scored by other teams per season
other_hteams_groupby= other_hteams.groupby('Season').sum()
other_ateams_groupby= other_ateams.groupby('Season').sum()


#other teams goals
other_ateams_goals=other_ateams_groupby.tail(other_ateams_groupby.shape[0]-1)
other_hteams_goals=other_hteams_groupby.tail(other_hteams_groupby.shape[0]-1)
# other_hteams_goals
# other_ateams_goals

#renaming the other teams goals to reduce confusion when I merge the columns
# other_hteamg_rename=other_hteams_goals.rename(columns={'FT_home_goals':'other_FT_home_goals'}) 
# other_ateamg_rename= other_ateams_goals.rename(columns={'FT_away_goals':'other_FT_away_goals'})
```

```python
goal_comparison= {'BIG6 Total Home Goals':homet_season_goals['FT_home_goals'],
                  'Other Teams Total Home Goals':other_hteams_goals['FT_home_goals'],
                  'BIG6 Total Away Goals':awayt_season_goals['FT_away_goals'],
                 'Other Teams Total Away Goals':other_ateams_goals['FT_away_goals']}

goal_comparison_df= pd.DataFrame(data=goal_comparison)
goal_comparison_df['Total Home Goals']= goal_comparison_df['BIG6 Total Home Goals']+goal_comparison_df['Other Teams Total Home Goals']
goal_comparison_df['Total Away Goals']= goal_comparison_df['BIG6 Total Away Goals']+goal_comparison_df['Other Teams Total Away Goals']
goal_comparison_df['Home ratio']=goal_comparison_df['BIG6 Total Home Goals']/goal_comparison_df['Total Home Goals']
goal_comparison_df['Away ratio']=goal_comparison_df['BIG6 Total Away Goals']/goal_comparison_df['Total Away Goals']
goal_comparison_df['Total Goal Ratio']=(goal_comparison_df['BIG6 Total Away Goals']+goal_comparison_df['BIG6 Total Home Goals'])/(goal_comparison_df['Total Away Goals']+goal_comparison_df['Total Home Goals'])
```
